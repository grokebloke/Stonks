```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import nltk

nltk.download("vader_lexicon")
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyser = SentimentIntensityAnalyzer()

def calc_sentiment_score(text):
    return analyser.polarity_scores(text)["compound"]
```

```python
df = pd.read_csv("../Scraped News/META_news.csv")
df.head()

print(df["snippet"][4])
print(analyser.polarity_scores(df["snippet"][4]))

analyser = SentimentIntensityAnalyzer()
df["sentiment_score"] = df["snippet"].apply(calc_sentiment_score)

df["sentiment_score"].plot(
    kind="bar", figsize=(10, 5), title="Sentiment Score of " + "META"
)

df.to_csv("../Sentiment Scores/META_news_sentiment.csv")
```
